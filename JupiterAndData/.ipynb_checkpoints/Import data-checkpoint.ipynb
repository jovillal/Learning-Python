Now we are going to geocode (transform an address to a coordinates). For this geopy is needed, and also an internet connection.

```python
import geopy
dir(geopy)
```

The idea is to use a geocoder (like ArcGIS) to query a database with the address and get the coordinates.

```python
from geopy.geocoders import ArcGIS
```

```python
nom = ArcGIS()
```

```python
nom.geocode('3666 21st St, San Fransisco, CA 94114')
```

```python
n = nom.geocode('Calle 104A #11B-45, Bogota, 110111')
```

To get the latitude and longitud use methods of the same name.

```python
[n.latitude,n.longitude]
```

Let us get the coordinates for df1

```python
df1
```

```python
df1["Address"] = df1["Address"] + ', ' + df1["City"] + ', ' + df1["State"] + ', ' + df1["Country"] 
```

```python
df1
```

Now apply the geocoder to get the location object

```python
df1["Locations"] = df1["Address"].apply(nom.geocode)
```

```python
df1
```

Now we extranc lattitude and longitude in new colums (careful with None values)

```python
df1["Latitude"] = df1["Locations"].apply(lambda x: x.latitude if x != None else None)
```

```python
df1
```

```python
df1["Longitude"] = df1["Locations"].apply(lambda x: x.longitude if x != None else None)
```

```python
df1
```
